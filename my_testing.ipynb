```python
import os

def create_ln_files(folder_path="read2016/lines"):
    # Lists to hold filenames
    train_list = []
    valid_list = []
    test_list = []

    # Scan the directory
    if not os.path.exists(folder_path):
        print(f"Error: Folder '{folder_path}' not found.")
        return

    print(f"Scanning {folder_path}...")
    files = os.listdir(folder_path)

    for filename in files:
        # We only care about images, not .txt files
        if filename.endswith(".jpeg") or filename.endswith(".jpg") or filename.endswith(".png"):
            if filename.startswith("train"):
                train_list.append(filename)
            elif filename.startswith("valid"):
                valid_list.append(filename)
            elif filename.startswith("test"):
                test_list.append(filename)

    # Write to .ln files
    with open("read2016/train.ln", "w") as f:
        f.write("\n".join(train_list))
    with open("read2016/val.ln", "w") as f:
        f.write("\n".join(valid_list))
    with open("read2016/test.ln", "w") as f:
        f.write("\n".join(test_list))

    print(f"Done! Found {len(train_list)} train, {len(valid_list)} val, {len(test_list)} test.")

if __name__ == "__main__":
    create_ln_files()
```
