```python
import torch
import matplotlib.pyplot as plt
import os
import sys
import numpy as np

# Import your project modules
from model import HTR_VT
from data import dataset
from utils import utils

# --- ⚙️ USER CONFIGURATION ⚙️ ---
# 1. The folder name inside ./output/
EXP_NAME = "test_run"  

# 2. Image dimensions (Must match training settings)
IMG_WIDTH = 512
IMG_HEIGHT = 64

# 3. Paths to your data
DATA_PATH = "./data/read2016/lines/"
TRAIN_LIST = "./data/read2016/train.ln"
TEST_LIST = "./data/read2016/test.ln"
# -------------------------------

def demo():
    # --- 1. SETUP ---
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"✅ Running on Device: {device}")

    class Args: pass
    args = Args()
    args.img_size = [IMG_WIDTH, IMG_HEIGHT]
    args.data_path = DATA_PATH
    
    # --- 2. LOAD DATASETS ---
    print("⏳ Loading dataset definitions...")
    
    if not os.path.exists(TRAIN_LIST):
        print(f"❌ Error: Could not find {TRAIN_LIST}")
        return

    train_ds = dataset.myLoadDS(TRAIN_LIST, DATA_PATH, args.img_size)
    ralph = train_ds.ralph
    
    # Create the Label Converter
    converter = utils.CTCLabelConverter(ralph.values())

    # Load Test Dataset
    test_ds = dataset.myLoadDS(TEST_LIST, DATA_PATH, args.img_size, ralph=ralph)
    
    # --- 3. LOAD MODEL ---
    print("🏗️  Building model...")
    
    # --- FIX: FORCE NB_CLS TO 90 TO MATCH CHECKPOINT ---
    # The saved model expects 90 outputs.
    model = HTR_VT.create_model(nb_cls=90, img_size=args.img_size[::-1])
    
    checkpoint_path = f'./output/{EXP_NAME}/best_CER.pth'
    if not os.path.exists(checkpoint_path):
        print(f"❌ Error: Model weights not found at {checkpoint_path}")
        return

    print(f"💾 Loading weights from {checkpoint_path}...")
    checkpoint = torch.load(checkpoint_path, map_location='cpu')
    
    # Fix dictionary keys (remove 'module.' prefix)
    state_dict = {}
    for k, v in checkpoint['state_dict_ema'].items():
        new_key = k.replace('module.', '')
        state_dict[new_key] = v
        
    model.load_state_dict(state_dict)
    model.to(device)
    model.eval()

    # --- 4. RUN INFERENCE ---
    print("🚀 Running prediction on the first 5 images...")
    
    loader = torch.utils.data.DataLoader(test_ds, batch_size=1, shuffle=False)
    
    images_to_show = []
    labels_to_show = []
    preds_to_show = []
    
    with torch.no_grad():
        for i, (img_tensor, label_str) in enumerate(loader):
            if i >= 5: break 
            
            img_tensor = img_tensor.to(device)
            
            # Forward pass
            preds = model(img_tensor)
            
            # Decode probabilities to Text
            preds_size = torch.IntTensor([preds.size(1)])
            _, preds_index = preds.max(2)
            preds_index = preds_index.transpose(1, 0).contiguous().view(-1)
            pred_str = converter.decode(preds_index.data, preds_size.data)[0]
            
            # Store results
            images_to_show.append(img_tensor.cpu())
            labels_to_show.append(label_str[0])
            preds_to_show.append(pred_str)

    # --- 5. VISUALIZE ---
    print("🎨 Generating results image...")
    fig, axes = plt.subplots(len(images_to_show), 1, figsize=(10, 3 * len(images_to_show)))
    if len(images_to_show) == 1: axes = [axes] 

    for ax, img, true_lbl, pred_lbl in zip(axes, images_to_show, labels_to_show, preds_to_show):
        # Convert tensor back to image
        np_img = img.squeeze(0).squeeze(0).numpy()
        np_img = (np_img - np_img.min()) / (np_img.max() - np_img.min())
        
        ax.imshow(np_img, cmap='gray')
        ax.set_title(f"Truth: {true_lbl}\nPred : {pred_lbl}", fontsize=14, color="blue", loc='left')
        ax.axis('off')

    plt.tight_layout()
    output_filename = f"results_{EXP_NAME}.png"
    plt.savefig(output_filename)
    print(f"✅ Success! Image saved to: {output_filename}")

if __name__ == "__main__":
    demo()
```
